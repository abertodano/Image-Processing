```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

for i in range(33370): #for all frames
    if i%100 == 0: #only look at some frames
        image = cv2.imread(f'frames/frame{i}.jpg')
        template = cv2.imread('Subimages/chicken.jpg')
        heat_map = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)

        if np.max(heat_map) > 0.7:
            h, w, _ = template.shape
            y, x = np.unravel_index(np.argmax(heat_map), heat_map.shape)
            cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 5)
            print(f'Frame number: {i}')
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
