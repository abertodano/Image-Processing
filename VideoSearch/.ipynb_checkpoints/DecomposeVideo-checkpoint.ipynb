```python
#IMPORTS
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# GET FRAMES FROM VIDEO

```python
vidcap = cv2.VideoCapture('video.mp4')
success,image = vidcap.read()
count = 0
while success:
    if count%10 == 0: #only write 1 in 10 frames
        cv2.imwrite("frames/frame%d.jpg" % count, image)     # save frame as JPEG file      
    success,image = vidcap.read()
    #print('Read a new frame: ', success)
    count += 1
```

# MAKE SOBEL CONVOLUTIONS OUT OF FRAMES

```python
for i in range(33370): #for all frames
    if i%20 == 0: #only look at kept frames #3270, 25600
        image = cv2.imread(f'frames/frame{i}.jpg') #load image
        #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert to greyscale
        
        sobel_horizontal = cv2.Sobel(image, cv2.CV_32F, 1, 0, ksize=3) #get horizontal and vertical sobel derivatives
        sobel_vertical = cv2.Sobel(image, cv2.CV_32F, 0, 1, ksize=3)
        sobel = np.sqrt(sobel_horizontal**2 + sobel_vertical**2)/600 #RMS and normalize. 1/400-1/600 seems optimal? idk
        
        sobelimage = np.minimum(cv2.cvtColor(sobel, cv2.COLOR_RGB2GRAY),1)*256 #convert full color sobel to greyscale, cap at 1, multiply by pixel value 256
        cv2.imwrite("sobelframes/frame%d.jpg" % i, sobelimage) #write
```

```python
plt.imshow(np.minimum(cv2.cvtColor(sobel, cv2.COLOR_RGB2GRAY),1))
```

# FIND SUBIMAGES

```python
#get subimages to search and filepaths
figures = ['fox','shark','chicken','parrot','jellyfish'] #list of figures to look for
filepathsdict = {_:'subimages/' + _ + 'loutline.jpg' for _ in figures} #list of filepaths to of subimage template image files
```

```python
for i in range(33370): #for all frames
    if i%1000 == 0: #only look at some frames
        image = cv2.imread(f'sobelframes/frame{i}.jpg') #load the sobel image
        for figure in figures: #for each figure we are searching for
            template = cv2.imread(filepathsdict[figure]) #get the corresponding template
            heat_map = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED) #heat map its inner product with image
        
            if np.max(heat_map) > 0.2: #if above a certain threshold
                h, w, _ = template.shape
                y, x = np.unravel_index(np.argmax(heat_map), heat_map.shape)
                cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 5) #draw a big red rectangle where the match was found
                
                print(f'Found match for {figure} at frame number: {i}')
                cv2.imwrite(f'foundframes/{figure}_frame{i}.jpg', image) #write the found frame
```
